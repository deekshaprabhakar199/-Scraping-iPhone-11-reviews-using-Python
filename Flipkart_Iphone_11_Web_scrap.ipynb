```
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```
title=[]
rating=[]
review=[]
Customer=[]
location=[]
review_likes=[]
review_dislikes=[]
rev_like_only=[]
loc_only=[]
url='https://www.flipkart.com/apple-iphone-11-pro-silver-64-gb/product-reviews/itmaf0c50869f4e6?pid=MOBFKCTSZJCKYJ55&sortOrder=MOST_HELPFUL&certifiedBuyer=false&aid=overall&page='
for i in range(1,15):
  page=requests.get(url+str(i))
  soup= BeautifulSoup(page.content,'html.parser')
  ### For Extracting all titles
  ##title= soup.select('p._2xg6Ul')
  title.append(soup.select('p._2xg6Ul'))
  ### For Extracting all reviews
  ##rating= soup.select('div.hGSR34.E_uFuv')
  rating.append(soup.select('div.hGSR34.E_uFuv'))
  ### For Extracting all reviews
  ##review= soup.select('div.qwjRop div div')
  review.append(soup.select('div.qwjRop div div'))
  ## For Extracting all the customer names 
  ##Customer= soup.select('p._3LYOAd._3sxSiS')
  Customer.append(soup.select('p._3LYOAd._3sxSiS'))
  ### For Extracting only the locations
  location= soup.select('p._19inI8 span')
  for i in range(0, len(location)):
    if i%2 != 0:
      loc_only.append(location[i])
  ### For Extracting likes on all the reviews
  ##review_likes= soup.select('span._1_BQL8')
  ##review_likes.append(soup.select('span._1_BQL8'))
  review_likes= soup.select('div._2ZibVB span._1_BQL8')
  for i in range(0,len(review_likes)):
    if i%2 ==0:
      rev_like_only.append(review_likes[i])
  ### For Extracting dislikes on all the reviews
  ##review_dislikes= soup.select('span._1_BQL8')
  review_dislikes.append(soup.select('div._2ZibVB._1FP7V7 span._1_BQL8'))


```
